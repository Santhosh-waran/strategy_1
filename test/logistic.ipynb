```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns 
%matplotlib inline
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
# from sklearn.datasets import  load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv("EURUSD.csv", parse_dates=[0], index_col=0)
df = df.Close.to_frame()
```

```python
df["Return"] = np.log(df.Close.div(df.Close.shift(1)))
# df["Return"] = df.Close.div(df.Close.shift(1))
df["direction"] = np.sign(df.Return)
```

```python
for i in range(6):
    df[f"lag_{i}"] = df.Return.shift(i)
```

```python
df.dropna(inplace=True)
df.head(4)
```

```python
x= df.drop("direction",axis=1)
y = df.direction
```

```python
model = LogisticRegression()
model.fit(x,y)
```

```python
df["pred"] = model.predict(x)
```

```python
df.pred.value_counts()
```

```python
hit = np.sign(df.Return*df.pred).value_counts()
hit
```

```python
hit_ratio = hit[1] / sum(hit)
hit_ratio
```
